```python
# матрица смежности

R = [
[0,1,4,1],
[1,0,3,2],
[4,3,0,0],
[1,2,0,0]]

print(R)
# матрица расстояний
D = [
    [0, 1, 2, 3],
    [1, 0, 1, 2],
    [2, 1, 0, 1],
    [3, 2, 1, 0]
]

n = 4
dot = lambda x, y: sum(i * j for i, j in zip(x, y))
ls = n * [-1]
```

```python
from itertools import permutations


def calc(ls):
    result = 0 
    for i in range(len(ls)):
        sm = 0
        for j in range(len(ls)):
            sm += R[ls[i]][ls[j]] * D[i][j]
        result += sm
    return result


def mn():
    return min(calc(el) for el in list(permutations([0, 1, 2, 3], 4)))
```

```python
def w(ls):
    result = 0
    for i in range(n):
        if i in ls:
            a = []
            b = []
            for j in range(n):
                if j not in ls:
                    a.append(R[i][j])
                if ls[j] == -1:
                    b.append(D[ls.index(i)][j])
            result += dot(sorted(a, reverse=True), sorted(b))
    return result


def v(ls):
    a = []
    b = []
    for i in range(n):
        for j in range(n):
            if i not in ls and j > i and j not in ls:
                a.append(R[i][j])
            if ls[i] == -1 and j > i and ls[j] == -1:
                b.append(D[i][j])
    return dot(sorted(a, reverse=True), sorted(b))


def f(ls):
    result = 0 
    for i in range(n):
        if i in ls:
            for j in range(n):
                if j in ls and j > i:
                    result += R[i][j] * D[ls.index(i)][ls.index(j)]
    return result


def F(ls):
    a = f(ls)
    b = w(ls)
    c = v(ls)
    return f(ls) + w(ls) + v(ls)
```

```python
print('Нижняя граница для исходных данных:', F(ls), '\n')

for i in range(n):
    sp = [1000] * n
    for j in range(n):
        if ls[j] == -1:
            new_ls = ls.copy()
            print(f'Помещаем элемент e{i + 1} на позицию p{j + 1}')
            new_ls[j] = i
            f_ = f(new_ls)
            print(f'Получим F(q) = {f_}')
            w_ = w(new_ls)
            print(f'Получим w(P) = {w_}')
            v_ = v(new_ls)
            print(f'Получим v(P) = {v_}')
            print(f'Таким образом, нижняя граница F(P) = {f_} + {w_} + {v_} = {f_ + w_ + v_}')
            sp[j] = f_ + w_ + v_ 
    print('Ярус нижних границ:', sp)
    k = min(list(range(len(sp))), key=lambda x: sp[x])
    print(f'Назначаем элемент e{i + 1} на позицию p{k + 1}', '\n')
    ls[k] = i
print('Полученное размещение:', *['{' + str(i + 1) + '}' for i in ls])
print('Итого, суммарный вес размещения:', calc(ls))
print('Минимальное размещение через полный перебор:', mn())
```
